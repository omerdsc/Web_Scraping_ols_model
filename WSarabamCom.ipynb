```python
from bs4 import BeautifulSoup as bts
import pandas as pd
import requests
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result=requests.get(url,headers={"User-Agent":"Chrome/137.0.7151.104"})
    soup=bts(result.text,"html.parser")
    return soup
```

```python
pages=["https://www.arabam.com/ikinci-el/otomobil?take=50"]
for page in range(2,50):
    pages.append("https://www.arabam.com/ikinci-el/otomobil?take=50&page="+str(page))
pages
```

```python
products=[]

for page in pages:
    html=getAndParseURL(page)
    for product in html.findAll("td",{"class":"listing-modelname pr"}):
        products.append("https://www.arabam.com/"+product.a.get("href"))
products
```

```python
all_features = []

for link in products:
    html = getAndParseURL(link)

    features = {}

    container = html.find("div", {"class": "product-properties-container"})
    if container:
        items = container.find_all("div", {"class": "property-item"})
        for item in items:
            key = item.find("div", {"class": "property-key"})
            value = item.find("div", {"class": "property-value"})
            if key and value:
                features[key.get_text(strip=True)] = value.get_text(strip=True)

    price_tag = html.find("div", {"data-testid": "desktop-information-price"})
    if price_tag:
        features["Fiyat"] = price_tag.get_text(strip=True)

    all_features.append(features)

df = pd.DataFrame(all_features)
df.head()
```
