### Kilometre içerinin temizlenmesi

```python
df["Kilometre"] = (
    df["Kilometre"]
    .str.replace("km", "", regex=False)
    .str.replace(" ", "", regex=False)
    .str.replace(".", "", regex=False)
    .astype(int)
)
```

```python
def temizle_motor_hacmi(deger):
    if pd.isna(deger):
        return np.nan
    if "-" in deger:
        # Aralık varsa
        sayilar = [int(s) for s in re.findall(r'\d+', deger)]
        if len(sayilar) == 2:
            return sum(sayilar) / 2
        else:
            return np.nan
    else:
        # Tek sayı varsa
        match = re.search(r'\d+', deger)
        if match:
            return int(match.group())
        else:
            return np.nan

df["Motor Hacmi"] = df["Motor Hacmi"].apply(temizle_motor_hacmi)
```

```python
def temizle_motor_gücü(deger):
    if pd.isna(deger) or deger == "-":
        return np.nan
    
    # Aralık var mı?
    if "-" in deger:
        sayilar = [int(s) for s in re.findall(r'\d+', deger)]
        if len(sayilar) == 2:
            return sum(sayilar) / 2
        else:
            return np.nan
    
    # Tek sayı var
    match = re.search(r'\d+', deger)
    if match:
        return int(match.group())
    
    return np.nan

df["Motor Gücü"] = df["Motor Gücü"].apply(temizle_motor_gücü)
```

```python
def temizle_model_motor(deger):
    if pd.isna(deger):
        return np.nan
    # Mesela "1.6 TDI" içinde baştaki ondalıklı sayı
    match = re.search(r'\d+\.\d+', deger)
    if match:
        return float(match.group())
    # Ondalık yok ama tam sayı mı var? "6.6 V12" gibi
    match = re.search(r'\d+', deger)
    if match:
        return float(match.group())
    return np.nan

df["Model"] = df["Model"].apply(temizle_model_motor)
```

```python
df["Ort. Yakıt Tüketimi"] = (
    df["Ort. Yakıt Tüketimi"]
    .astype(str)                        
    .str.replace("lt", "", regex=False) 
    .str.replace(" ", "", regex=False)  
    .str.replace(",", ".", regex=False) 
    .replace("nan", np.nan)            
    .astype(float)                      
)
```

```python
df["Yakıt Deposu"]=(df["Yakıt Deposu"].astype(str)
                   .str.replace("lt","")
                   .str.replace(" ","")
                   .replace("nan",np.nan)
                   .astype(float))
```

```python
import re
import numpy as np

def temizle_ekspertiz(deger):
    if pd.isna(deger) or deger in ["Belirtilmemiş"]:
        return pd.Series([np.nan, np.nan])
    if deger == "Tamamı orjinal":
        return pd.Series([0, 0])
    degisen = re.search(r'(\d+) değişen', deger)
    boyali = re.search(r'(\d+) boyalı', deger)
    d = int(degisen.group(1)) if degisen else 0
    b = int(boyali.group(1)) if boyali else 0
    return pd.Series([d, b])

df[["Degisen", "Boyalı"]] = df["Boya-değişen"].apply(temizle_ekspertiz)
```

## Boş sütünları doldurma


Model boş sütünlarını doldurma. Yanmalı motorlara en çok tekrar eden ama elektrikli araçlarıda boş bırak

```python
# En çok geçen Model değerini bul
en_cok = df["Model"].mode()[0]

# Koşullu doldurma:
df.loc[(df["Model"].isna()) & (df["Yakıt Tipi"] == "Elektrik"), "Model"] = 0
df.loc[(df["Model"].isna()) & (df["Yakıt Tipi"] != "Elektrik"), "Model"] = en_cok
```

```python
# 1️⃣ En çok geçen Motor Gücü (mode) bulunur
en_cok_mg = df["Motor Gücü"].mode()[0]

# 2️⃣ Elektrikli olanlara 0, diğer eksiklere mode atanır
df.loc[(df["Motor Gücü"].isna()) & (df["Yakıt Tipi"] == "Elektrik"), "Motor Gücü"] = 0
df.loc[(df["Motor Gücü"].isna()) & (df["Yakıt Tipi"] != "Elektrik"), "Motor Gücü"] = en_cok_mg
```

## Bağımlı ve Bağımsız Değişkenler

```python
y=df["Fiyat"]
x=df.drop("Fiyat",axis=1)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Random Forest Modeli

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=100,  # Ağaç sayısı
    random_state=42,
    max_depth=None,  # İsteğe bağlı sınırlarsın
    n_jobs=-1  # Tüm çekirdekleri kullan
)

model.fit(x_train, y_train)
```

## Tahmin & Skor

```python
y_pred = model.predict(x_test)

from sklearn.metrics import r2_score, mean_squared_error

print("R² Skoru:", r2_score(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))
print(model.score(x_test,y_test))
```
